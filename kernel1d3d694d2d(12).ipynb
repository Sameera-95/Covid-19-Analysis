# Extract Details about china

```python
df9 = df1.drop(df1[~ ( df1['Country/Region'].str.startswith("China") ) ].index)
df9 = df9.reset_index()
df9.head()
```

```python
df10 = []
df11 = []
df12 = []
for i in range(len(df9.columns)-5):
    Total = df9.iloc[:,i+5].sum()
    df10.append(Total)
    df11.append(pd.to_datetime(df9.columns[i+5]))
    df12.append(pd.to_datetime(df9.columns[i+5]) - df11[0])
```

```python
plt.rcParams["figure.figsize"] = (15,10)

ax = plt.subplot(111)
ax.plot(df11, df10,'x', color = 'red')
ax.set_title('Number of reported cases in China', fontsize = '14')
ax.set_xlabel('Date', fontsize = '14')
ax.set_ylabel('Total cases',  fontsize = '14')
ax.xaxis_date()

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.yaxis.set_ticks_position('left')
    
plt.show()
```

```python
df12 = pd.DataFrame(df12)
df14 = df12.astype('timedelta64[h]')/24
```

```python
plt.rcParams["figure.figsize"] = (15,10)

ax = plt.subplot(111)
ax.plot(df14, df10,'x', color = 'red', label="Data")
ax.set_title('Number of reported cases in China', fontsize = '14')
ax.set_xlabel('Days', fontsize = '14')
ax.set_ylabel('Total cases',  fontsize = '14')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.yaxis.set_ticks_position('left')

mymodel = np.poly1d(np.polyfit(df14[0], df10, 6))
myline = np.linspace(1, 80, 100)

plt.plot(myline, mymodel(myline), label="Ploynimial fit")
plt.legend(loc="upper left", fontsize = '14')
plt.show()

from sklearn.metrics import r2_score
print(r2_score( df10, mymodel(df14[0]))) 
```

```python
for k in range(10):
    mymodel = np.poly1d(np.polyfit(df14[0], df10, k))
    print(r2_score( df10, mymodel(df14[0]))) 
```
