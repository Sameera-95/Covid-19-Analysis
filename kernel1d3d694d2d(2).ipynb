```python
df1 = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/time_series_covid_19_confirmed.csv')
df1.head()
```

```python
df1.dtypes
```

```python
df2 = df1.iloc[225,4:]
df2.head
```

```python
df2 = pd.DataFrame(df2)
df2.head()
```

```python
df3 = df2.reset_index()
df3.columns = ['date','count']
df3.head()
```

```python
df3['date'].to_string()
df3.dtypes
```

```python
df3['date']= pd.to_datetime(df3['date']) 
df3.dtypes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (15,10)

ax = plt.subplot(111)
ax.bar(df3['date'], df3['count'], width=1)
ax.set_ylim(0, max(df3['count'])+50000)
ax.xaxis_date()
plt.show()
```
